### Breast Cancer Data Analysis

```python
from sklearn.datasets import load_breast_cancer
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load the data
data = load_breast_cancer().data
```

```python
print('Shape of the dataset: ', data.shape)
```

```python
#create a dataset
dataset = pd.DataFrame(data)
dataset.head()
```

```python
features = load_breast_cancer().feature_names
features
```

```python
dataset.columns = features
dataset['Label']=np.reshape(load_breast_cancer().target,(569,1))
dataset.head()
```

### Data Visualisation

#### Normalisation/standardisation of data

```python
# Transformation into normally distributed data: mean = 0, std = 1
from sklearn.preprocessing import StandardScaler
x=dataset.loc[:,features].values
x= StandardScaler().fit_transform(x)
x.shape
```

```python
#check the normalisation of data
print('Mean: ', np.mean(x))
print('Standard deviation: ', np.std(x))
```

```python
normalised_data = pd.DataFrame(x, columns = features)
normalised_data.head()
 
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(x)
```

```python
principal = pd.DataFrame(data=principal_components, columns=['pc1','pc2'])
```

```python
principal.head()
```

```python
print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))
```

```python
#Plot
ax1=principal.plot.scatter(x='pc1', y ='pc2', c='DarkBlue')
```
